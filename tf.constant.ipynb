```python
# Importing Tensorflow
import tensorflow as tf
print(tf.__version__)
```

```python
#create Tensor with tf.constant()
scaler = tf.constant(7)
scaler
```

```python
#Check The Number Of Diamension Of the scaler .(ndim stand for number of diamension)
scaler.ndim
```

```python
# Crete A Vector
vector = tf.constant([10 , 2])
vector
```

```python
#Number Diamenion of vector
vector.ndim
```

```python
#Create A Matrix (Matrix Is Minimum 2 daimensional )
matrix = tf.constant([[10,7],
                     [12,13]])
matrix
```

```python
# Number of diamension of matrix
matrix.ndim
```

```python
#Create Another matrix
another_matrix = tf.constant([[100,200],
                              [400,500],
                              [400,800]], dtype = tf.float16)
another_matrix
```

```python
#Crerate A tensor
tensor = tf.constant([[[10,20,30],
                       [40,50,60]],
                      [[20,30,40],
                       [20,30,40]],
                      [[50,60,70],
                       [90,80,70]]])
tensor
```

```python
#Number of diamension of tensor
tensor.ndim
```
