```python
 y = tf.constant(np.random.randint(0,100 , size = 50))
```

```python
y
```

```python
#Minimum of the tensor
tf.reduce_min(y)
```

```python
#Maximum of the tensor
tf.reduce_max(y)
```

```python
# Mean of the Tensor
tf.reduce_mean(y)
```

```python
# Sum of the Tensor
tf.reduce_sum(y)
```

## Mathmatical Operation In Tensorflow ##

```python
 s = tf.range(1,20)
```

```python
s
```

```python
tf.square(s)
```

```python
#Square Root Try to normal method and see its give Error or not
tf.sqrt(s)
```

```python
# Now try to right Method for Square root
tf.sqrt(tf.cast(s, dtype = tf.float32))
```

```python
# Now try To Normal Method For log Operation
tf.log(s)
```

```python
# Now To Right Method For log Operation
tf.math.log(tf.cast(s, dtype = tf.float32))
```
