
Matrix Multification
In Machine Learning Matrix Multification one of the most common tensor  operation
There Are Two RuLes our Tensors( Or metrics) need to fullfill if are going to matrix multify them
1.The Inner diamenion Must match
2.The resulting matrix has the shape of the inner diamensiom

```python
import tensorflow as tf
```

```python
tensor = tf.constant([[10,7],
                      [3,4]])
tensor
```

```python
tf.matmul(tensor,tensor)
```

```python
tensor,tensor
```

```python
# Matrix multification in Python
tensor @ tensor
```

```python
tensor*tensor
```

```python
# Creating [3,2] matrix
x = tf.constant([[1,2],
                 [2,4],
                 [6,8]])
y = tf.constant([[2,3],
                 [5,7],
                 [1,9]])
```

```python
x,y
```

```python
x @ y
```

```python
tf.matmul(x,y)
```

```python
# lets change the Y shape
tf.reshape(y, shape=(2,3))
```

```python
x.shape , tf.reshape(y, shape=(2,3)).shape
```

```python
x @ tf.reshape(y, shape=(2,3))
```

```python
x.shape ,tf.reshape(x, shape=(2,3)).shape , y.shape
```

```python
 # Change The Shape of the X Metrics and multification-
 tf.reshape(x, shape=(2,3)) @ y
```

```python
tf.matmul(tf.reshape(x, shape=(2,3)), y)
```

```python
# Now Use transpose
x,tf.transpose(x)
```

```python
tf.transpose(x)@ y
```

```python
tf.matmul(tf.transpose(x),y)
```

```python
x, y
```

```python
tf.tensordot(  x , tf.transpose(y), axes =1)
```

```python
tf.tensordot(x, tf.reshape(y , shape=(2,3)), axes = 1)
```

```python
 tf.matmul(x , tf.transpose(y))
```

```python
tf.matmul(x, tf.reshape(y , shape=(2,3)))
```

```python
tf.tensordot(tf.transpose(x),y , axes=1)
```

```python
tf.tensordot(tf.reshape(x , shape=(2,3)),y,axes = 1)
```

```python
# The Information about the metrics
print('Normal Y :--------------')
print('\n', y)
print('Reshape Multification X: -------------')
print('\n', tf.matmul(tf.reshape(x, shape=(2,3)),y))
print('Transpose Multification X:-------------')
print('\n', tf.matmul(tf.transpose(x),y))
print('Reshape Multification Y :-------')
print('\n', tf.matmul(x, tf.reshape(y, shape=(2,3))))

print('Transpose Multification Y :----------------')
print('\n',tf.matmul(x, tf.transpose(y)))

```
