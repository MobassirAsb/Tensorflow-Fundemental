```python
import tensorflow as tf
```

```python
#Create Changable tensor
changable_tensor = tf.Variable([10,7])
changable_tensor
```

```python
#Create Unchangable tensor
unchangable_tensor = tf.constant([10,7])
unchangable_tensor
```

```python
changable_tensor[0].assign(7)
```

```python
#Unchangable tensor(tf.constant always be constant. it unchangable )
unchangable_tensor[0].assign(7)
```
